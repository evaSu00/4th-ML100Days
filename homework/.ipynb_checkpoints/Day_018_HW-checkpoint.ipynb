```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
%matplotlib inline
plt.style.use('ggplot')

# 設定 data_path
dir_data = './data/'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

## 練習時間
參考 Day 17 範例程式，離散化你覺得有興趣的欄位，並嘗試找出有趣的訊息

```python
app_train['DAYS_BIRTH'] = app_train['DAYS_BIRTH']/365
```

```python
print("Correlation between TARGET and DAYS_BIRTH", app_train['DAYS_BIRTH'].corr(app_train['TARGET']))
(app_train['DAYS_BIRTH']).describe()
```

```python
# 將年齡相關資料, 另外存成一個 DataFrame 來處理
age_data = app_train[['TARGET', 'DAYS_BIRTH']]
age_data
```

```python
# 新增欄位，20-70切11分
bins_cut = np.linspace(20,70,num=11)
age_data["DAYS_BIRTH_cut"] = pd.cut(age_data['DAYS_BIRTH'], bins = bins_cut)
print(age_data["DAYS_BIRTH_cut"].value_counts())
age_data.head()
```

```python
age_data_group = age_data.groupby('DAYS_BIRTH_cut').mean()
age_data_group
```

```python
plt.figure(figsize = (8, 8))

# 以年齡區間為 x, target 為 y 繪製 barplot
"""
Your Code Here
"""
px = age_data_group.index
py = age_data_group['TARGET']
sns.barplot(px, py)

# Plot labeling
plt.xticks(rotation = 75); plt.xlabel('Age (years)'); plt.ylabel('Target (%)')
plt.title('Failure to Repay by Age Group');
```

```python
#可以非常明顯地看出趨勢，越年輕的客戶族群的違約遲繳的可能越大！
```

```python
#找最大最小值
app_train['AMT_INCOME_TOTAL'].describe()
```

```python
app_train_dic = {
    'SK_ID_CURR' : app_train['SK_ID_CURR'],
    'TARGET' : app_train['TARGET'],
    'AMT_INCOME_TOTAL' : app_train['AMT_INCOME_TOTAL']
}

app_train_df = pd.DataFrame(app_train_dic)
```

```python
app_train_df
```

```python
# 新增欄位，25650-9000000切11分
bins_cut = np.linspace(25650,9000000,num=11)
app_train_df["Income_total_cut"] = pd.cut(app_train_df['AMT_INCOME_TOTAL'], bins = bins_cut)
print(app_train_df["Income_total_cut"].value_counts())
app_train_df.head()
```

```python
# 新增欄位 "Income_total_qcut", 對收入做等頻劃分
app_train_df["Income_total_qcut"] = pd.qcut(app_train_df['AMT_INCOME_TOTAL'], 5)
print(app_train_df["Income_total_qcut"].value_counts())
app_train_df.head()
```

```python
# 計算Income_total_cut的 Target,Income的平均值
Income_total_cut_groups  = app_train_df.groupby('Income_total_cut').mean()
Income_total_cut_groups
```

```python
Income_total_cut_groups.index
```

```python
plt.figure(figsize = (8, 8))

# 以年齡區間為 x, target 為 y 繪製 barplot
"""
Your Code Here
"""
px = Income_total_cut_groups.index
py = Income_total_cut_groups['TARGET']
sns.barplot(px, py)

# Plot labeling
plt.xticks(rotation = 75); plt.xlabel('Income_total (cut)'); plt.ylabel('Target (%)')
#plt.title('Failure to Repay by Age Group');
```

```python
# 計算Income_total_qcut的 Target、 AMT_INCOME_TOTAL的平均值
Income_total_qcut_groups  = app_train_df.groupby('Income_total_qcut').mean()
Income_total_qcut_groups
```

```python
plt.figure(figsize = (8, 8))

# 以年齡區間為 x, target 為 y 繪製 barplot
"""
Your Code Here
"""
px = Income_total_qcut_groups.index
py = Income_total_qcut_groups['TARGET']
sns.barplot(px, py)

# Plot labeling
plt.xticks(rotation = 75); plt.xlabel('Income_total (Qcut)'); plt.ylabel('Target (%)')
#plt.title('Failure to Repay by Age Group');
```
